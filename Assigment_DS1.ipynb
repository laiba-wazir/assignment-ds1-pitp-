How to access data?

```python
# there rae two ways to access data from dictionaray
dict2= {
    "1":  "corolla",
    "2":  "farrari",
    "3":  "alto",               # 6 keys and its 6 values
    "4":  "altis",
    "5":   "cultus",
    "6":   "city"

}

a = dict2["2"]       # method 1: use square bracket
b = dict2.get("6")   # method 2: use get() function
a, b
```

values(): return all the values of the dictionary

```python
dict2.values()   # fetch all the values of "dict2" variable by using values() function
```

Change and update()

```python
dict2["4"] = "mango"    # use this key assignment for single changes
dict2                   # display the changed dictionary "dict2"
```

```python
dict2.update({"4": "Mercedes", "8": "Lamborghini"})    # use update() function for change or update multiple values at once
dict2                                                  # display the updated dictionary "dict2"
```

REMOVE ITEMS:

```python
# pop()
dict2.pop("3")        # pop() function removes the item with the specified key name
dict2                 # display
```

```python
# popitems()
dict2.popitem()      # popitem() function removes the last inserted item
dict2                # display
```

```python
# del keyword
del dict2["5"]     # del keyword removes the item with the specified key name
dict2              # display
```

```python
# clear()
dict1.clear()     # clear() function empties the dictionary.
dict1             # display
```

Nested Dictionary

```python
myfamily = {                   # defining dictionary "myfamily"
    "child1": {                # the key "child1" has values in dictionary (nested dictionary)
        "name": "Zuha",
        "year": 2009
    },
    "child2": {                # the key "child2" has values in dictionary (nested dictionary)
        "name": "Mohammad Ahmed",
        "year": 2014
    },
    "child3": {                # the key "child3" has values in dictionary (nested dictionary)
        "name": "Hasan",
        "year": 2017
    }
}

myfamily
```
